```python
from faker import Faker
import random
import pandas as pd

fake = Faker()
Faker.seed(42)
random.seed(42)

num_students = 200   # you can change this later

# --- Students Table ---
students = []
for i in range(1, num_students + 1):
    students.append({
        "student_id": i,
        "name": fake.name(),
        "age": random.randint(18, 25),
        "gender": random.choice(["Male", "Female", "Other"]),
        "email": fake.email(),
        "phone": fake.phone_number(),
        "enrollment_year": random.choice([2019, 2020, 2021, 2022]),
        "course_batch": random.choice(["Batch A", "Batch B", "Batch C"]),
        "city": fake.city(),
        "graduation_year": random.choice([2023, 2024, 2025])
    })
students_df = pd.DataFrame(students)

# --- Programming Table ---
programming = []
for sid in range(1, num_students + 1):
    programming.append({
        "programming_id": sid,
        "student_id": sid,
        "language": random.choice(["Python", "SQL", "Java"]),
        "problems_solved": random.randint(10, 300),
        "assessments_completed": random.randint(1, 10),
        "mini_projects": random.randint(0, 5),
        "certifications_earned": random.randint(0, 3),
        "latest_project_score": random.randint(40, 100)
    })
programming_df = pd.DataFrame(programming)

# --- Soft Skills Table ---
soft_skills = []
for sid in range(1, num_students + 1):
    soft_skills.append({
        "soft_skill_id": sid,
        "student_id": sid,
        "communication": random.randint(40, 100),
        "teamwork": random.randint(40, 100),
        "presentation": random.randint(40, 100),
        "leadership": random.randint(40, 100),
        "critical_thinking": random.randint(40, 100),
        "interpersonal_skills": random.randint(40, 100)
    })
soft_skills_df = pd.DataFrame(soft_skills)

# --- Placements Table ---
placements = []
for sid in range(1, num_students + 1):
    placements.append({
        "placement_id": sid,
        "student_id": sid,
        "mock_interview_score": random.randint(30, 100),
        "internships_completed": random.randint(0, 3),
        "placement_status": random.choice(["Ready", "Not Ready", "Placed"]),
        "company_name": random.choice([None, "TCS", "Infosys", "Wipro", "Amazon", "Google"]),
        "placement_package": random.choice([None, 400000, 600000, 1200000, 2000000]),
        "interview_rounds_cleared": random.randint(0, 5),
        "placement_date": fake.date_this_year() if random.random() > 0.7 else None
    })
placements_df = pd.DataFrame(placements)

# Save to CSVs
students_df.to_csv("students.csv", index=False)
programming_df.to_csv("programming.csv", index=False)
soft_skills_df.to_csv("soft_skills.csv", index=False)
placements_df.to_csv("placements.csv", index=False)

print(" Dataset generated and saved as CSVs with full required schema")
```

```python
import mysql.connector
import pandas as pd
import numpy as np

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Abby@123",
    database="students_data"
)
cursor = conn.cursor()

# --- Function to clean NaN ---
def clean_row(row):
    return [None if (pd.isna(x) or x == "nan") else x for x in row]

# --- Insert Students ---
students_df = pd.read_csv("students.csv")
for _, row in students_df.iterrows():
    cursor.execute("""
        INSERT INTO students (student_id, name, age, gender, email, phone, enrollment_year, course_batch, city, graduation_year)
        VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
    """, clean_row(row))
conn.commit()

# --- Insert Programming ---
programming_df = pd.read_csv("programming.csv")
for _, row in programming_df.iterrows():
    cursor.execute("""
        INSERT INTO programming (programming_id, student_id, language, problems_solved, assessments_completed, mini_projects, certifications_earned, latest_project_score)
        VALUES (%s,%s,%s,%s,%s,%s,%s,%s)
    """, clean_row(row))
conn.commit()

# --- Insert Soft Skills ---
soft_skills_df = pd.read_csv("soft_skills.csv")
for _, row in soft_skills_df.iterrows():
    cursor.execute("""
        INSERT INTO soft_skills (soft_skill_id, student_id, communication, teamwork, presentation, leadership, critical_thinking, interpersonal_skills)
        VALUES (%s,%s,%s,%s,%s,%s,%s,%s)
    """, clean_row(row))
conn.commit()

# --- Insert Placements ---
placements_df = pd.read_csv("placements.csv")
for _, row in placements_df.iterrows():
    cursor.execute("""
        INSERT INTO placements (placement_id, student_id, mock_interview_score, internships_completed, placement_status, company_name, placement_package, interview_rounds_cleared, placement_date)
        VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s)
    """, clean_row(row))
conn.commit()

print(" All data inserted successfully")

cursor.close()
conn.close()
```
